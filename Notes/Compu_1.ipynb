## Monomial Order

**1.2/(2.1)** Monomial Ordering is the ordering on $k[x_1,x_2,...,x_n]$ s.t.
1. total and well order (can be changed with $\forall \alpha: \alpha\geq 0$)
2. $x^{\alpha}>x^{\beta}\implies x^{\alpha+\gamma}>x^{\beta+\gamma}$

e.g.
1. **2.2/def3** Lexicographic Order 
2. **2.2/def5** Graded Lexicographic Order: first graded (by total degree), then lexicographic.
3. **2.2/def6** Graded Lexicographic Order: first graded (by total degree), then lexicographic reversely.

## Monomial Ideal and Groebner basis

**2.4/def1** Monomial Ideal is the ideal generated by monomials. 

**2.5/def5** The subset $G\subset I\subset k[x_1,x_2,...,x_n]$ is a Groebner (standard) basis if $$<LT(g_1),LT(g_2),...,LT(g_t)>=<LT(I)>=<LM(I)>$$ (which is actually monomial ideal). 

It's minimal if in addition s.t. $$\forall p\in G: LC(p)=1 \quad \text{and} \quad LT(p)\notin <LT(G-\{p\})>.$$

It's reduced (**unique** and stronger than minimal!) if in addition s.t. $$\forall p\in G: LC(p)=1 \quad  \text{and} \quad \text{all monomials of p}\notin <LT(G-\{p\})>.$$

Remark:
1. Actually we can fulfill the **ideal equality algorithm** through the uniqueness of reduced Groebner basis.

**2.4/Thm5(Dickson's Lemma)** For any monomial ideal $I=<x^{\alpha}:\alpha\in A>=<x^{\alpha_i}:\{\alpha_{1},...,\alpha_{n}\}\in A>$.

## Elimination to Extention

**3.1/Thm2(Elimination)** Let $I\in k[x_1,x_2,...,x_n]$ and G a Groebner basis w.r.t lex order $x_1>x_2>...>x_n$, then
$$ G_l = G \cap k[x_1,...,x_n] \text{ is a Groebner basis of } I_l=I\cap k[x_{l+1},...,x_n]$$

**3.1/Thm3(Extension)** Let $I=\{f_1,f_2,...,f_s\}\in \mathbb{C}[x_1,..,x_n]$ and write 
$$f_i = g_i x_1^{N_i}+ \text{terms where $x_1$'s degree < $N_1$}$$

Then
$$ (a_2,a_3,...,a_n) \notin V(g_1,g_2,...,g_s) \implies (a_2,a_3,...,a_n) \in \pi_1(V(I))$$

**Remark:** 
1. $\mathbb{C}$ in **Extension Theorem** can be changed to arbitrary $k$ s.t. $\bar{k}=k$.

**Proof** Consider $\pi: \mathbb{C}[x_1,..,x_n]\to \mathbb{C}[x_1]$ and try $\pi(I)\neq \mathbb{C}[x_1].$ (Apply Resultant Technique)

Here's the corollary of extension theorem.

**4.1/Thm1(Weak Null-stellen-satz)** $\bar{k}=k$ and $I\in k[x_1,x_2,..,x_n]$, then
$$V(I)=\emptyset \implies I=k[x_1,x_2,..,x_n].$$
